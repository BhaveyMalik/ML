```python
from nltk.stem import WordNetLemmatizer
from tqdm import tqdm
import re
```

```python
# cleaning data
not_used_words = ['www', 'youtube', 'com', 'http', 'jpg', 'vimeo', 'facebook', 'twitter', 'tweet', 'whatsapp']
pattern = r'[0-9]'

def clear_text(data):
    data_length=[]
    lemmatizer = WordNetLemmatizer()
    cleaned_text=[]
    for sentence in tqdm(data):
        sentence = sentence.lower()

        #removing links from text data
        sentence = re.sub('https?://[^s<>]+www\/[^\s<>]+', ' ', sentence)

        #removing other symbols
        sentence = re.sub('[^0-9a-z]', ' ', sentence)

        #further removing not used words
        for words in not_used_words: 
            sentence = re.sub(words, ' ', sentence)
            
        # removing redundant personalities
        for words in personality_types:
            words.lower()
            sentence = re.sub(words, ' ', sentence)
        
        data_length.append(len(sentence.split()))
        
        #removing numbers
        sentence = re.sub(pattern, ' ', sentence)
        
        #clearing extra spaces
        sentence = re.sub('\s+', ' ', sentence)
        
        
        #appending in cleaned sentences
        cleaned_text.append(sentence)
        
    return cleaned_text,data_length
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#gives most frequent words
def most_freq_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    
    #coutning occurences
    occurence = bag_of_words.sum(axis=0) 
    words_frequency = [(word, occurence[0, idx]) for word, idx in vec.vocabulary_.items()]
    
    #sorting occurences
    words_frequency =sorted(words_frequency, key = lambda x: x[1], reverse=True)
    
    #returning top n frequent words
    return words_frequency[:n]
```
